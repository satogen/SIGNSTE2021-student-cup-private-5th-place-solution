```python
# Earth Engine Python APIのインポート
import ee
# GEEの認証・初期化
ee.Authenticate()
ee.Initialize()
from google.colab import drive
drive.mount('/content/drive')
```

```python
import numpy as np
```

```python
world_data = ee.FeatureCollection("users/halogen322/world_countries")
```

```python
class CFG:
  target_satellite =  'MODIS/006/MOD11A1'
  target_band = 'LST_Day_1km'
  start_year = 2001
  end_year = 2021
  target_col = ["NAME", "NAME_ID","mean"]
  target_polygon = world_data
  filen_name = "temp_" 

  PRODUCTION = False
  OUT_PUTPATH = 'GEE_temp_data'
```

```python
reducers = ee.Reducer.mean();

def ntl(image):
  pas = image.reduceRegions(
    collection=CFG.target_polygon,  
    reducer=reducers,  
    scale=500);
  return pas;



def export_table(table, description):
  task = ee.batch.Export.table.toDrive(**{ 
    'collection': ee.FeatureCollection(table), 
    'description': description,
    'folder':CFG.OUT_PUTPATH, 
    'selectors':(CFG.target_col),
  })
  if CFG.PRODUCTION:
    task.start()
  print(f'export {description} file complete!')
```

```python
CFG.PRODUCTION = True
```

```python
for year in np.arange(CFG.start_year, CFG.end_year, 1, dtype='int'):
  print(str(year) + ' year start!')

  for month in np.arange(1, 13, 1):
    print(str(year) + '/' + str(month))

    # 月ごとの値を算出
    filter_range = ee.Filter.date(
      str(year) + '-' + str(month) + '-01', 
      (str(year) + '-' + str(month + 1) if (month != 12) else str(year+ 1) + '-' + '1')  + '-01'
    )

    # 日付を決定する
    dataset = ee.ImageCollection(CFG.target_satellite).filter(filter_range)

    # 対象バンドの選択
    surfaceAirTemperature = dataset.select(CFG.target_band)
    now_img = ntl(surfaceAirTemperature.mean())
    
    # 月のデータをダウンロード
    export_table(now_img, CFG.filen_name + str(year) + '-' + str(month));
```
